### Section 1. Basic Concepts

__print()__

A function that simply prints something in the screen. In below snippet, computer prints 'Hello world!' in the screen.

```python
print('Hello world!') 
```

The content must be surrounded by a pair of quotation marks. Both single and double quotation marks are acceptable. 

```python
print("Hello world double quotation!")
print('Hello world single quotation!')
```

Special symbols in the content need to be "skipped" by a slash (\). An example is shown in below snippet.

```python
print('I\'m a student')
```

Actually, in above example we may use a pair of double quotation marks to enclose the content, and the prime (') does not need to be skipped in this case, as the computer realizes the prime is not part of the quotation marks.

```python
print("I'm a student")
```

__input()__

A function that takes user input from the console. In below snippet, the computer ***pauses*** and waits for user to type something and then assigns it to the variable 'user_name1'. 

```python
user_name1 = input()
```

```python
user_name2 = input('What is your name? ') # Computer pauses and waits for user input.
print('Hello,', user_name2)
```

__Variables__ 

A variable has a name and can be used to hold values. 

A variable name must:
1. start with a letter or underscore (_),
2. contains only alpha-numeric characters and underscores,
3. not be Pyhon keywords, such as 'if', 'else', 'while', etc.

Examples of acceptable variable names: 'a_number', '_number', 'number1'.
Examples of unaccetable variable names: '1_number', 'number$', 'if'.

Variable names are case sensitive. For example, "a_number" and "A_Number" are two different variables.

There is no strict requirement for variable naming convention. It is preferred to use the "snake case" style, e.g., lowercase workds connected by under scores. For example, "a_good_variable_name".

__Data Types__

In Python, the most common data types include __integer__, __float__ (decimal number), __string__, and __boolean__(True or False). We don't have to tell the computer what data type a variable is. Python will decide based on the value we assign to this variable. In below snippet, the four variables are integer, float, string, and boolean (True of False), respectively.

```python
an_integer_number = 1
a_float_number = 1.0
a_string = "John is smart."
a_boolean = True
```

In below snippet, we print out the types of these variables. Note function __type()__ is used to check the type of variables.

```python
print(type(an_integer_number))
print(type(a_float_number))
print(type(a_string))
print(type(a_boolean))
```

Note that Python determines the data type of a variable based on the __last__ value it assigns to this variable. For example in below snippet, variable "greeting" is first assigned a string but then an integer number. Therefore, "greeting" is an integer.

```python
greeting = "Hellow world"
greeting = 1234
print(type(greeting))
```

__Comments__

In Python, we can add some comments, which is only to help people, including yourself, to understand the program. A comment starts with "#". A comment will not be executed by Python. An example is shown as below

```python
age = 19
# Check if age is no less than 18.
print(age >= 18)
```

__Numerical Representations__

1) Underscores. For big numbers, i.e., 12000300, we can add underscores to help read it. For example as below

```python
# underscores in numbers
a_big_number = 12_000_300
print(a_big_number)
```

2) Scientific notations. We can use "E" or "e" to denote "10 powers of". For example "3E3" is "3 times 10 powers of 3".

```python
x = 3e3
print(x)

y = 3e-3
print(y)
```

3) Octal numbers. Start with "0O" (number zero and capital "O") or "0o" (number zero and lowercase "o")

```python
print(0o123)
```

4) Hexadecimal numbers. Start with "0X" (number zero and capital "X") or "0x" (number zero and lowercase "x")

```python
print(0X12F)
```

__Operators__


```python
# plus
print(2 + 3) # 5

# minus
print(2 - 3) # -1

# multiplition
print(2 * 3) # 6

# standdard division
print(5 / 2) # 2.5

# integer division: returns the whole number only
print(5 // 2) # 2

# modulus division: returns the remainder
print(5 % 2) # 1

# power operator
print(2 ** 3) # 8 
```

In addition, subtraction, and multiplition operations, if both two numbers are integers, the result is integer. If at least one number is a float, the result will be a float.

```python
print(2 * 3.0) # 6.0
```

```python
print(2 * 3) # 6
```

__Reassigning values__

Note that in Python, a variable can be re-assigned values, even of different types.

```python
age = 28
print("Age:", age)

age = 29.5
print("New Age:", age)
```


Also try to understand below operations.

```python
age = 28
print(age)

age += 1 # 'age" is reassigned a new value of 'age + 1' (29)
print(age)

age *= 2 # 'age' is reassigned a new value of 'age * 2' (58)
print(age)

age /= 2 # 'age' becomes (58 / 2 = 29.0). Note division results in a float number
print(age)
```
